```python
# Check for duplicates
duplicates = data.duplicated().sum()
print(f'Number of duplicate rows: {duplicates}')

# If duplicates exist, you might want to drop them
# data = data.drop_duplicates()

# Verify data types
print(data.dtypes)

# Calculate descriptive statistics
print(data.describe())

# Calculate quantiles only for numeric columns
numeric_data = data.select_dtypes(include=['number'])  # Selects only numeric columns
Q1 = numeric_data.quantile(0.25)
Q3 = numeric_data.quantile(0.75)
IQR = Q3 - Q1

# Identifying outliers
outliers = ((numeric_data < (Q1 - 1.5 * IQR)) | (numeric_data > (Q3 + 1.5 * IQR))).any(axis=1)
print(f'Number of outlier rows: {outliers.sum()}')

# Displaying outliers (optional)
print(data[outliers])

```
